```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv(r'https://drive.google.com/uc?export=download&id=1xxDtrZKfuWQfl-6KA9XEd_eatitNPnkB')
df.head()
```

```python
X = df.drop('price', axis=1)
y = df['price']
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=51)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
sc.fit(X_train)
```

```python
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
# Predict the value
X_test[0,:]
```

```python
lr.predict([X_test[0,:]])
```

```python
lr.predict(X_test)
```

```python
y_test
```

```python
lr.score(X_test, y_test)
```

```python
# Implementing Ridge & Lasso Regression
```

```python
from sklearn.linear_model import Ridge, Lasso
```

```python
rd = Ridge()
rd.fit(X_train, y_train)
rd.score(X_test, y_test)
```

```python
ls = Lasso()
ls.fit(X_train, y_train)
(ls.score(X_test, y_test)*100).round(2)
```

```python
rd2 = Ridge(alpha=2)
rd2.fit(X_train, y_train)
rd2.score(X_test, y_test)
```

```python
ls2 = Lasso(alpha=2)
ls2.fit(X_train, y_train)
(ls2.score(X_test, y_test)*100).round(2)
```

```python
ls3 = Lasso(alpha=3)
ls3.fit(X_train, y_train)
(ls3.score(X_test, y_test)*100).round(2)
```

```python
# Model Evaluation
```

```python
# Root Mean Square

y_pred = lr.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
rmse
```

```python
from sklearn.metrics import r2_score
```

```python
rs = r2_score(y_test, y_pred)
rs
```

```python
# Polynomial Linear Regression
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures(degree=2)
pf.fit(X_train)
X_train_poly = pf.transform(X_train)
X_test_poly = pf.transform(X_test)
```

```python
print(X_train_poly.shape)
print(X_test_poly.shape)
```

```python
lr = LinearRegression()
lr.fit(X_train_poly, y_train)
```

```python
lr.score(X_test_poly, y_test)
```

```python
lr.predict([X_test_poly[2,:]])
```

```python
y_test
```

```python
y_pred = lr.predict(X_test_poly)
y_pred
```

```python
mse = mean_squared_error(y_test,y_pred)
mse
```

```python
rmse = np.sqrt(mse)
rmse
```
